```python
from GaNN.models.GaNN import GaNN
import torch 
import numpy as np 
from GaNN.models.HyperNet import HyperNet
from GaNN.models.HyperNet2 import HyperNet as HyperNet2
from GaNN.models.MLP import MLP

import time 

%load_ext autoreload
%autoreload 2
```

```python
x = torch.randn(10,2)
y =  torch.randn(10,2)

model3 = HyperNet2(MLP(in_channels=2, out_channels=2, hidden_channels=10, layers=2), stochastic_channels=10, width=10)
_ = model3(x)
```

```python
x = torch.randn(10,2)
y =  torch.randn(10,2)

model = HyperNet(MLP(in_channels=2, out_channels=2, hidden_channels=10, layers=2), stochastic_channels=10, width=10)
model2 = GaNN(x.size(1), y.size(1), hidden_channels=10, layers=2, gaussian_channels=10, width=10)
model3 = HyperNet2(MLP(in_channels=2, out_channels=2, hidden_channels=10, layers=2), stochastic_channels=10, width=10)


t1 = [] 
t2 = [] 
t3 = []
for i in range(100): 
    tic = time.time() 
    model(x, samples=1000)
    t1.append(time.time() - tic)

    tic = time.time() 
    model2(x, samples=1000)
    t2.append(time.time() - tic)

    tic = time.time() 
    model3(x, samples=1000)
    t3.append(time.time() - tic)
```

```python
print('wrapper hnet:', np.mean(t1))
print('from scratch hnet:', np.mean(t2))
print('vectorized hnet:', np.mean(t3))
```

```python
# just testing... 
optim = torch.optim.Adam(model.parameters())
out = model(x).mean()
out.backward()
optim.step()
```
