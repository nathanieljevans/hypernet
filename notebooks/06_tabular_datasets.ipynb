# Diabetes Dataset 

```python
for loss_ in ['mse', 'nll', 'edl']:
    print('Loss objective:', loss_)
    
    train_kwargs = {'num_epochs' : 1000,
                        'nsamples' : 1000,
                        'batch_size' : 250,
                        'lr' : 1e-4,
                        'loss_fn':loss_,
                        'use_cuda':True}
        
    mlp_kwargs = {'hidden_channels':50, 
                    'layers':2,
                    'dropout':0.}

    hnet_kwargs = {'stochastic_channels':25, 
                   'width':25}

    print(run_sklearn_dataset('diabetes', 'hnet', mlp_kwargs=mlp_kwargs, hnet_kwargs=hnet_kwargs, train_kwargs=train_kwargs))
    print('---')
```

## Monte Carlo Dropout 

```python
train_kwargs = {'num_epochs' : 1000,
                    'batch_size' : 250,
                    'lr' : 1e-3,
                    'loss_fn':'mse',
                    'use_cuda':True}
    
model_kwargs = {'hidden_channels':50, 
                'layers':2,
                'dropout':0.5}

ien_test_dict = run_sklearn_dataset('diabetes', 'mcdo', model_kwargs, train_kwargs)
ien_test_dict
```
